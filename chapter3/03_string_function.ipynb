```python
# Basic String Functions in Python

text = "  hello python programming 123  "

print("Original String:")
print(text)

# 1. lower()
print("\n1. lower():")
print(text.lower())

# 2. upper()
print("\n2. upper():")
print(text.upper())

# 3. title()
print("\n3. title():")
print(text.title())

# 4. capitalize()
print("\n4. capitalize():")
print(text.capitalize())

# 5. strip()
print("\n5. strip():")
print(text.strip())

# 6. replace()
print("\n6. replace():")
print(text.replace("python", "java"))

# 7. find()
print("\n7. find():")
print(text.find("python"))

# 8. count()
print("\n8. count():")
print(text.count("o"))

# 9. startswith()
print("\n9. startswith():")
print(text.strip().startswith("hello"))

# 10. endswith()
print("\n10. endswith():")
print(text.strip().endswith("123"))

# 11. split()
print("\n11. split():")
words = text.split()
print(words)

# 12. join()
print("\n12. join():")
joined_text = "-".join(words)
print(joined_text)

# 13. isalpha()
print("\n13. isalpha():")
print("Python".isalpha())

# 14. isdigit()
print("\n14. isdigit():")
print("12345".isdigit())

# 15. isalnum()
print("\n15. isalnum():")
print("Python123".isalnum())

# 16. len()
print("\n16. len():")
print(len(text))

# 17. String Slicing
print("\n17. String Slicing:")
sample = "Python"

print("First character:", sample[0])
print("First 3 characters:", sample[0:3])
print("Last character:", sample[-1])

# Final Cleaned Example
print("\nFinal Cleaned Example:")

name = "   shreya gade   "

clean_name = name.strip().title()

print(clean_name)
```
