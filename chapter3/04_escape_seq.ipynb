```python
# Combined Program of Escape Sequences in Python

print("===== ESCAPE SEQUENCES IN PYTHON =====")

# 1. New Line (\n)
print("\n1. New Line Example:")
print("Hello\nPython")

# 2. Tab Space (\t)
print("\n2. Tab Space Example:")
print("Name\tAge\tCity")
print("Shreya\t18\tPune")

# 3. Backslash (\\)
print("\n3. Backslash Example:")
print("C:\\Users\\Shreya\\Documents")

# 4. Single Quote (\')
print("\n4. Single Quote Example:")
print('It\'s a Python Program')

# 5. Double Quote (\")
print("\n5. Double Quote Example:")
print("He said \"Python is easy\"")

# 6. Backspace (\b)
print("\n6. Backspace Example:")
print("Helloo\b")

# 7. Carriage Return (\r)
print("\n7. Carriage Return Example:")
print("12345\rABC")

# 8. Form Feed (\f)
print("\n8. Form Feed Example:")
print("Hello\fWorld")

# 9. Vertical Tab (\v)
print("\n9. Vertical Tab Example:")
print("Hello\vWorld")

print("\n===== PROGRAM COMPLETED =====")
```
