```python
# Combined Program of Dictionary Methods in Python

print("===== DICTIONARY METHODS IN PYTHON =====")

# Creating Dictionary
student = {
    "name": "Shreya",
    "age": 18,
    "city": "Pune"
}

# Original Dictionary
print("\nOriginal Dictionary:")
print(student)

# 1. keys()
print("\n1. keys()")
print(student.keys())

# 2. values()
print("\n2. values()")
print(student.values())

# 3. items()
print("\n3. items()")
print(student.items())

# 4. get()
print("\n4. get()")
print(student.get("name"))

# 5. update()
print("\n5. update()")
student.update({"course": "Python"})
print(student)

# 6. pop()
print("\n6. pop()")
student.pop("age")
print(student)

# 7. popitem()
print("\n7. popitem()")
student.popitem()
print(student)

# 8. copy()
print("\n8. copy()")
new_student = student.copy()
print(new_student)

# 9. setdefault()
print("\n9. setdefault()")
student.setdefault("country", "India")
print(student)

# 10. clear()
print("\n10. clear()")
student.clear()
print(student)

print("\n===== PROGRAM COMPLETED =====")
```
